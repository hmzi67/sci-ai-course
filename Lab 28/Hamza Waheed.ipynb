## 1.	Lemmatization Basics
- Given the list of words ['running',       'better', 'mice', 'geese', 'happiest'], use the WordNetLemmatizer to lemmatize each word. What are the resulting lemmatized forms?


```python
words = ['running', 'better', 'mice', 'geese', 'happiest']
```

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
print([wnl.lemmatize(w) for w in words])
```

### 3.	Stemming with Porter Stemmer
- Given the text: "The quick brown fox jumps over the lazy dog", use the PorterStemmer to stem each word in the sentence. What are the resulting stemmed words?


```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer() 

text= "The quick brown fox jumps over the lazy dog"

words = word_tokenize(text)
words
```

```python
[ps.stem(w) for w in words]
```

### 4.	Comparing Stemming Methods
- Use both the PorterStemmer and LancasterStemmer on the word list ['flying', 'flies', 'flies', 'fly', 'flyer']. What are the differences in the outputs of the two stemmers?


```python
from nltk.stem import PorterStemmer, LancasterStemmer

ps = PorterStemmer()
ls = LancasterStemmer()

words4 = ['flying', 'flies', 'flies', 'fly', 'flyer']
```

```python
ps_words = [ps.stem(w) for w in words4]
ps_words
```

```python
ls_words = [ls.stem(w) for w in words4]
ls_words
```

### 6.	Part of Speech Tagging

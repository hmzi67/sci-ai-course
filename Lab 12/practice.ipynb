# 6.	Calculate Average Height:
### a.	What is the method to calculate the average height of students in the DataFrame? Write the code to compute and display this average.


```python
sumOfHeight = data['height'].values.sum()

average = sumOfHeight / len(data['height'])
```

```python
average.round()
```


# 7.	Group By and Aggregate:
### a.	How can you group the data by gender and calculate the average age for each gender? Write the code to perform this aggregation and print the results.


```python
average_of_genders = data.groupby('gender').agg('age').mean()
```

```python
average_of_genders
```

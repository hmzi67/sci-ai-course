**Built-in Data Types**

In programming, data type is an important concept.

Variables can store data of different types, and different types can do different things.

Python has the following data types built-in by default, in these categories:

Text Type:	str

Numeric Types:	int, float, complex

Sequence Types:	list, tuple

Mapping Type:	dict

Set Types:	set

Boolean Type:	bool

**Brief Description of Python Data Types**

Integer (int):

Represents whole numbers, both positive and negative, without a fractional component.

```python
age = 25
print(type(age))  
```

Float (float):

Represents real numbers with a decimal point (fractional part).

```python
price = 19.99
print(type(price))  
```

String (str):

A sequence of characters enclosed in single, double, or triple quotes.

```python
greeting = "Hello, World!"
print(type(greeting))  
```

Boolean (bool):

Represents truth values: True or False.

```python
is_active = True
print(type(is_active))  
```

List (list):

An ordered, mutable collection of items (can be of different data types).

```python
numbers = [1, 2, 3, 4]
print(type(numbers))  
```

Tuple (tuple):

An ordered, immutable collection of items.

```python
coordinates = (10, 20)
print(type(coordinates))  
```

Dictionary (dict):

A collection of key-value pairs, where each key is unique.

```python
student = {"name": "Alice", "age": 20}
print(type(student)) 
```

Set (set):

An unordered collection of unique items (no duplicates allowed).



```python
unique_numbers = {1, 2, 3, 3, 4}
print(type(unique_numbers))  
```

Complex (complex):

Represents complex numbers with real and imaginary parts.

```python
complex_num = 3 + 4j
print(type(complex_num))  
```

Casting in python is therefore done using constructor functions:

int() - constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number)

float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)

str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals

```python
#Integers:

x = int(1)   
y = int(2.8)
z = int("3") 
```

```python
#Floats:

x = float(1)     
y = float(2.8)   
z = float("3")   
w = float("4.2") 
```

```python
#Strings:

x = str("s1") 
y = str(2)   
z = str(3.0)  
```

# Algebric Equations and Graphs

**Finding the Slope and Intercept of a Line**

Python program that finds the slope and intercept of a line given two points on the line.

Use the input() function to get the coordinates of the two points from the user.

Use algebraic manipulation to find the slope and intercept.

```python
# Get the coordinates of the two points from the user
x1 = float(input("Enter the x-coordinate of the first point: "))
y1 = float(input("Enter the y-coordinate of the first point: "))
x2 = float(input("Enter the x-coordinate of the second point: "))
y2 = float(input("Enter the y-coordinate of the second point: "))

# Calculate the slope
m = (y2 - y1) / (x2 - x1)

# Calculate the intercept
b = y1 - m * x1

print("The slope is " + str(m) + " and the intercept is " + str(b))
```

**Graphing Linear Equations**

Python program that graphs a linear equation of the form y = mx + b, where m and b are constants.

Use the matplotlib library to create the graph.

Use the input() function to get the values of m and b from the user.

```python
import matplotlib.pyplot as plt

# Get the values of m and b from the user
m = float(input("Enter the value of m: "))
b = float(input("Enter the value of b: "))

# Create the graph
x = [i for i in range(-10, 11)]
y = [m*i + b for i in x]

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = " + str(m) + "x + " + str(b))
plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.show()
```

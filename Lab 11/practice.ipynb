```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
n = 10      # size
p = 0.5     # probablity
```

```python
# probablity distribution  at x = 5
st.binom.pmf(5, n, p)
```

```python
# Quantile at alpha = 0 - 1
st.binom.ppf(0.3, n, p)
```

```python
# P(0 <= x <= 5)
st.binom.cdf(5, n, p)
```

```python
st.binom.cdf(4, n, p) - st.binom.cdf(2, n, p)
```

```python
x = np.arange(0, 11)
plt.scatter(x, st.binom.pmf(x, n, p), color='orange')
plt.show()
```

#### Normal

```python
mu = 5
sigma = 2
```

```python
st.norm.pdf(0, loc=mu, scale=sigma)
```

```python
st.norm.ppf(0.5, loc=mu, scale=sigma)
```

```python
st.norm.ppf(0.01, loc=mu, scale= sigma)
```

```python
st.norm.ppf(0.99, loc=mu, scale=sigma)
```

```python
st.norm.cdf(5, loc=mu, scale=sigma)
```

```python
st.norm.cdf(9, loc=mu, scale=sigma) - st.norm.cdf(3, loc=mu, scale=sigma)
```

```python
st.norm.cdf(9, loc=mu, scale=sigma) - st.norm.cdf(1, loc=mu, scale=sigma)
```

```python
x = np.linspace(-10, 10, 1000)
mu = 0
sigma = 2
plt.plot(x, st.norm.pdf(x, loc=mu, scale=sigma), color='red')
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
sigma = 2
mu_1 = -1
mu_2 = 0
mu_3 = +1
plt.plot(x, st.norm.pdf(x, loc=mu_1, scale=sigma), color='red')
plt.plot(x, st.norm.pdf(x, loc=mu_2, scale=sigma), color='green')
plt.plot(x, st.norm.pdf(x, loc=mu_3, scale=sigma), color='orange')
plt.show()
```

```python
x = np.linspace(-10, +10, 1000)
sigma_1 = 1
sigma_2 = 2
sigma_3 = 3
mu = 0

plt.plot(x, st.norm.pdf(x, loc=mu, scale=sigma_1), color="green")
plt.plot(x, st.norm.pdf(x, loc=mu, scale=sigma_2), color="purple")
plt.plot(x, st.norm.pdf(x, loc=mu, scale=sigma_3), color="skyblue")
plt.show()
```

#### Chi Square

```python
df = 5
```

```python
st.chi2.pdf(1, df)
```

```python
st.chi2.ppf(0.5, df)
```

```python
st.chi2.ppf(0.99,df)
```

```python
st.chi2.cdf(5, df=10)
```

```python
st.chi2.cdf(10, df=5) - st.chi2.cdf(5, df=5)
```

```python
x=np.linspace(0,+15,1000)
plt.plot(x, st.chi2.pdf(x,df=4),color='red')         # df = 1
plt.plot(x, st.chi2.pdf(x,df=2),color='green')       # df = 5
plt.plot(x, st.chi2.pdf(x,df=1),color='blue')       # df = 10
plt.ylim([0,1])
plt.show()
```

## Part 2: Control Flow
### 3. For Loop and If-Else
- Write a Python program that prints all even numbers between 1 and 20 using a for loop.
- Use an if-else statement inside the loop to print whether the number is odd or even.

Using for loop between 1 to 20 

```python
# use for loop to iterate from 0 to 21 and it take step size of 2
for i in range(0, 21, 2):
    print(i)
```

```python
# use for loop to iterate from 0 to 21
for i in range(1, 21):
    # Use if condition to check if modulus of i is equal to zero then it print that number. 
    if (i % 2) == 0:
        print(f"Even: {i}")
    else:
        print(f"Odd: {i}")
```

## Part 5: Data Visualization with Matplotlib

### 8. Line Graph
- Generate x-values from 0 to 10, and y-values as the square of x.
- Plot a line graph of x vs y with labels for the x-axis and y-axis and a title for the graph.

```python
x = np.arange(0, 10)
y = np.square(x)
```

```python
x
```

```python
y 
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, color='orange')
plt.title('Plotting a line graph x vs y')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### 9. Bar Chart
- Create a list of five categories (e.g., "A", "B", "C", "D", "E") and their respectivevalues.
- Plot a bar chart using Matplotlib to represent these categories.

```python
categories = ["A", "B", "C", "D", "E"]
y = np.random.randint(1, 50, 5)
```

```python
plt.bar(categories, y, color=plt.cm.Set1(np.arange(len(categories))))
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart for categorical data')
plt.show()
```

### 10. Histogram

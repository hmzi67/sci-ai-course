```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df = pd.read_csv('data_Iris.csv')
data_df.head()
```

```python
for col in data_df.columns:
    sns.boxplot(x=data_df[col])
    plt.title(col)
    plt.show()
```

```python
for col in data_df:
    figure = plt.subplot()
    sns.histplot(data_df, x=col, bins=15)
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

mm_scaler = MinMaxScaler()
numerical_df = data_df.select_dtypes(include='number').columns

data_df[numerical_df] = mm_scaler.fit_transform(data_df[numerical_df])
data_df
```

```python
for col in data_df:
    figure = plt.subplot()
    sns.histplot(data_df, x=col, bins=15)
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
for col in data_df.columns:
    sns.boxplot(x=data_df[col])
    plt.title(col)
    plt.show()
```

```python
ss_scaler = StandardScaler()
data_df[numerical_df] = ss_scaler.fit_transform(data_df[numerical_df])
data_df
```

```python
for col in data_df:
    figure = plt.subplot()
    sns.histplot(data_df, x=col, bins=15)
    plt.title(f"Histogram of {col}")
    plt.show()


for col in data_df.columns:
    sns.boxplot(x=data_df[col])
    plt.title(col)
    plt.show()
```

```python
rs_scaler = RobustScaler()
data_df[numerical_df] = rs_scaler.fit_transform(data_df[numerical_df])
data_df
```

```python
for col in data_df:
    figure = plt.subplot()
    sns.histplot(data_df, x=col, bins=15)
    plt.title(f"Histogram of {col}")
    plt.show()


for col in data_df.columns:
    sns.boxplot(x=data_df[col])
    plt.title(col)
    plt.show()
```

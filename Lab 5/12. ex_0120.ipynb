```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sigmoid Function

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Synthetic data (independent variable)
x = np.linspace(-10, 10, 100)

# Applying sigmoid function to data
y = sigmoid(x)

# Plot
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')

plt.grid(True)
plt.show()
```

# Trignometric

```python
# Generate angles between 0 and 2 (0 to 360 degrees)
x = np.linspace(0, 2 * np.pi, 1000)

# Compute the tangent values
tan_y = np.tan(x)

# Plot the tangent function
plt.plot(x, tan_y)

# Limit the y-axis to range between -10 and 10 to avoid showing the spikes
plt.ylim(-10, 10)

# Add labels and title
plt.title('Tangent Function with Limited Y-Axis')
plt.xlabel('Angle (radians)')
plt.ylabel('Tangent(x)')
plt.grid(True)

plt.show()
```

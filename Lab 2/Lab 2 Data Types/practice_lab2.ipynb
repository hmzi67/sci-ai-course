# Additional Practice

```python
num = 3

student_dict = {
        "name": "",
        "age": 0,
        "cgpa": 0.00 
    }

while num != 0:
    userInput = input("Enter a number: ")

    try:
        if "." in userInput:
            value = float(userInput)
            student_dict["cgpa"] = value
        else:
            value = int(userInput)
            student_dict["age"] = value
    except:
        value = str(userInput)
        student_dict["name"] = value
    
    num -= 1

for key, i in student_dict.items():
    print(f"{key}: {i}")
```

```python
import matplotlib.pyplot as plt


x = [i for i in range(-10, 11)]
y = [2*i - 3 for i in x]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Graph of y = " + str(2) + "x - " + str(3))
plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color="black")
plt.show
```

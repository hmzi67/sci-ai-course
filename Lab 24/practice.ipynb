```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=pd.DataFrame(data=data.data, columns=data.feature_names)
iris
```

```python
feature=pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
data.target
```

```python
target=pd.DataFrame(data.target, columns=["species"])
```

```python
target
```

```python
iris = pd.concat([feature, target], axis=1)

iris.rename({"sepal length (cm)":"sepal_length", "sepal width (cm)":"sepal_width",
             "petal length (cm)":"petal_length", "petal width (cm)":"petal_width"}, axis=1, inplace=True)
iris.columns
```

```python
iris['species'] = iris.species.map(lambda x: data.target_names[x])
iris.head()
```

```python
iris.isna().sum()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris.groupby('species').size()
```

```python
def boxplot_iris(feature_name, dataset):
    i=1
    plt.figure(figsize=(11, 9))
    for col in feature_name:
        plt.subplot(2, 2, i)
        plt.axis('on')
        plt.tick_params(axis='both', left=True,
                        top=False, right=False,
                        bottom=True, labelleft=False,
                        labeltop=False, labelright=False,
                        labelbottom=False
                        )
        dataset[col].plot(kind='box', subplots=True, sharex=False)
        plt.title(col)
        i += 1
    plt.show()
```

```python
boxplot_iris(iris.columns[:-1], iris)
```

```python
def histogram_iris(feature_name, dataset):
    i=1
    plt.figure(figsize=(11, 2))
    for col in feature_name:
        plt.subplot(1, 4, i)
        plt.axis('on')
        plt.tick_params(axis='both', left=True,
                        top=False, right=False,
                        bottom=True, labelleft=False,
                        labeltop=False, labelright=False,
                        labelbottom=False
                        )
        dataset[col].plot(kind='hist', color="blue", subplots=True, grid=True)
        plt.title(col)
        i += 1
    plt.show()
```

```python
histogram_iris(iris.columns[:-1], iris)
```

```python
corr = iris.corr(numeric_only=True)
cmap = sns.diverging_palette(220, 20, as_cmap=True)
plt.figure(figsize=(11, 9))
sns.heatmap(corr, cmap=cmap, annot=True, vmax=1.0, vmin=-1.0, center=0, square=True, linewidths=.5, cbar_kws={"shrink":.5})
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.plot()
```

```python
def piechar_iris(feature_names, target, dataset):
    i=1
    plt.figure(figsize=(11, 9))
    for colName in [target]:
        labels = []; sizes=[];
        df = dataset.groupby(colName).size()
        for key in df.keys():
            labels.append(key)
            sizes.append(df[key])
        plt.subplot(2, 2, i)
        plt.axis('on')
        plt.tick_params(axis='both', left=True,
                        top=False, right=False,
                        bottom=True, labelleft=True,
                        labeltop=True, labelright=False,
                        labelbottom=False)
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
        plt.axis('equal')
        i += 1
    plt.show()
```

```python
piechar_iris(iris.columns[:-1], iris.species, iris)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(iris.iloc[:,:-1], iris.iloc[:,-1],
                                                  test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(
    criterion='gini',
    splitter='best',
    max_depth=None,
    min_samples_split=2,
    random_state=42, 
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    class_weight=None
)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score, KFold

cv = KFold(n_splits=10, shuffle=True, random_state=42)
results = cross_val_score(model, X_train, y_train, cv=cv)
fin_result = np.mean(results)
```

```python
for i, _ in enumerate(results):
    print("{}th cross validation score: {}".format(i, _))
print("\ncross validation final score: {}".format(fin_result))
```

```python
from sklearn.model_selection import StratifiedKFold

cv=StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
results=cross_val_score(model, X_train, y_train, cv=cv)
fin_result=np.mean(results)
for i, _ in enumerate(results):
    print("{}th cross validation score: {}".format(i, _))
```

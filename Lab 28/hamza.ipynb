```python
text = """I'm Hamza Waheed abbasi. I'm a Software developer.
I've 1 year of experience in Software Developer at Code Hunts.
Currently I'm pursuing Master in DATA SCIENCE from FAST NUCES.
And I've completed my Bacheor's in Computer Science from ARID AGRICULTURAL.
I've completed many projects based on web and android development.
"""
```

```python
from nltk.tokenize import word_tokenize

words = word_tokenize(text)
words
```

Lemmatization Basics

```python
from nltk.stem import WordNetLemmatizer


wnl = WordNetLemmatizer()

print([wnl.lemmatize(w) for w in words])
```

Lemmatization with Parts of Speech


```python
print([wnl.lemmatize(w, 'v') for w in words])
```

Stemming with Porter Stemmer

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

print([ps.stem(w) for w in words])
```

verb

```python
print([ps.stem(w, 'v') for w in words])
```

adjective

```python
print([ps.stem(w, 'a') for w in words])
```

LancasterStemmer

```python
from nltk.stem import LancasterStemmer

l= LancasterStemmer()

[l.stem(w) for w in words]
```

```python
import nltk

pos_tagging = nltk.pos_tag(words)
pos_tagging
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

Loading dataset

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=data, x=data['sepal length (cm)'], edgecolor='black', hue='species')
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

Create a boxplot to compare the sepal width for each species. Which species has the
largest median sepal width?

```python
sns.boxplot(data, x='sepal width (cm)', hue='species')
plt.title('Sepal width for each Species')
plt.show()
```

Printing the highest median of sepal width

```python
species_mediam = data.groupby('species').median()
high_median = species_mediam['sepal width (cm)'].idxmax()
print("Heighest median of sepal width is: "+ high_median)
```

Generate a scatter plot to examine the relationship between petal length and petal width,

with points colored by species. Based on the scatter plot, how do petal length and petal
width vary across the three species?

```python
sns.scatterplot(data, x='petal length (cm)', y='petal width (cm)', hue='species')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Patel Length')
plt.ylabel('Patel Width')
plt.show()
```

Setosa:

This species is easily distinguishable as it clusters at the lower end with shorter petal lengths and smaller petal widths.
Petal length and width are relatively low and show little variation.


Versicolor:

This species lies in the mid-range of both petal length and width.
There is a moderate positive correlation between petal length and width.


Virginica:

Virginica occupies the higher range, with the longest petal lengths and largest petal widths.
Like Versicolor, the petal dimensions have a strong positive correlation.

Use Seaborn’s pairplot to visualize pairwise relationships between all numerical
features, colored by species.

```python
sns.pairplot(data, x_vars=data.select_dtypes(include=['number']), hue='species')
plt.suptitle('Pairplot of Iris Dataset Features', y=1.02) 
plt.show()
```

Create a heatmap to show the correlation between the numerical features. What insights
can you gather from the heatmap about the correlation between the sepal length and the
petal length?

```python
sns.heatmap(data.select_dtypes(include=['number']).corr(),  annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between the numerical features')
plt.show()
```

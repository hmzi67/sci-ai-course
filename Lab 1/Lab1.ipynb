**Program to check if a number is prime using if-else.**

```python
# Check if a number is prime
num = int(input("Enter a number: "))

if num > 1:  # Prime numbers are greater than 1
    for i in range(2, num):
        if num % i == 0:  # If divisible by any number other than 1 and itself
            print(num, "is not a prime number.")
            break
    else:  # This else belongs to the for loop, not the if
        print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")  # Numbers less than or equal to 1 are not prime
```

```python
num = int (input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, " is not a prime number")
            break
        else:
            print(num, " is a prime number.")
else:
    print(num, " is not a prime number.")
```

**Task for Students:**

Create a Program to Calculate the Factorial of a Number
        
        Objective:
Write a Python program to calculate the factorial of a given number entered by the user.

Instructions:
Input: The program should ask the user to input a non-negative integer.
Logic: Implement a loop to calculate the factorial of the number.
Output: The program should display the factorial of the given number.

```python
user_input = int(input("Enter a number: "))

fact = 1

if user_input < 0:
    print("Sorry, Factorial does not exist")
elif user_input == 0:
    print ("Factorial of 0 is 1")
else: 
    for i in range(1, user_input + 1):
        fact = fact * i
    print(f"factorial of {user_input} is {fact}")
```

**7. Error and Exception Handling**

```python
# Simple try-except block
try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter valid numbers.")
finally:
    print("Execution completed.")
```

```python
try:
    number1 = int(input("Enter a number: "))
    number2 = int(input("Enter a number: "))

    result = number1 / number2

    print("Result: ", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed")
except ValueError:
    print("Error: Invalid input. Please enter a valid numbers")
finally:
    print("Execution Completed")
```

**Explanation:**

The try-except block catches and handles exceptions.
ZeroDivisionError and ValueError are common exceptions.
The finally block runs regardless of exceptions.

**Task 1: Simple Calculator**

    Objective: Create a simple calculator that can perform addition, subtraction, multiplication, and division.

Instructions:

Prompt the user to enter two numbers.

Ask the user to choose an operation (addition, subtraction, multiplication, division (Choose operation (+, -, *, /)) ).

Perform the operation and display the result.

```python
def sum(a, b):
    return a + b
def sub (a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    return a / b

print(f"""1 - Addition
2 - Subtraction
3 - Multiplication
4 - Division
""")
user_inp = int(input("Select an option: "))

number_01 = int(input("Enter your first number: ")) 
number_02 = int(input("Enter your second number: ")) 

if user_inp == 1:
    result = sum(number_01, number_02)
elif user_inp == 2: 
    result = sub(number_01, number_02)
elif user_inp == 3: 
    result = multiply(number_01, number_02)
elif user_inp == 4: 
    result = divide(number_01, number_02)
else:
    print("You input invalid number. Try again")

print("Result: ", result)
```

**Task 3: Count Vowels in a String**

    Objective: Count the number of vowels in a given string.

Instructions:

Prompt the user to enter a string.

Use a loop to count the vowels (a, e, i, o, u) in the string.

Print the count.

```python
input_str = str(input("Enter a string: "))
print("The string is: ", input_str)

count = 0

for word in input_str:
    if word == 'a' or word == 'e' or word == 'i' or word == 'o' or word == 'u':
        count += 1
    

print("Number of vowels in the input string is ", count)
```

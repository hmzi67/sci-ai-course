```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
fig0 = plt.figure()
axes0 = fig0.add_axes([0, 0, 1, 1])
axes0.plot(x, y, color='red', linestyle='-') 
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('Line Plot')
plt.show()
```

```python
fig0 = plt.figure()
axes0 = fig0.add_axes([0, 0, 1, 1])
axes0.plot(x, y, color='red', linestyle='-') 
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('Line Plot')
axes0.set_xlim([0, 5])
axes0.set_ylim([-2, 2])
plt.show()
```

```python
fig0 = plt.figure(figsize=(5, 2), dpi=100)
axes0 = fig0.add_axes([0, 0, 1, 1])
axes0.plot(x, y, color='red', linestyle='--', label='Sin') 
axes0.plot(x, z, color='orange', linestyle='-.', label='Cos') 
axes0.legend(loc=0)
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('Line Plot')
plt.show()
```

```python
fig0.savefig('figure_01.png')
```

# TASK 2

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)
z = np.cos(x)

w = np.sin(x) + np.cos(x)

fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)
axs[0].plot(x, y, color='green', label='sin(x)')
axs[0].set_title("Sine y = sin(x)")
axs[0].set_ylabel("Y")
axs[0].grid(True)

axs[1].plot(x, z, color='red', label='cos(x)')
axs[1].set_title("Sine z = cos(x)")
axs[1].set_ylabel("Y")
axs[1].grid(True)

axs[2].plot(x, w, color='blue', label='sin(x) + cos(x)')
axs[2].set_title("Sine w = sin(x) + cos(x)")
axs[2].set_xlabel("X")
axs[2].set_ylabel("Y")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from matplotlib.gridspec import GridSpec

x = np.arange(0, 10, 0.1)
y = np.sin(x)
z = np.cos(x)

w = np.sin(x) + np.cos(x)

fig = plt.figure(figsize=(12, 5))
gs = GridSpec(1, 2, width_ratios=[2, 3])

axs1 = fig.add_subplot(gs[0, 0])
axs1.plot(x, y, color='green', label='sin(x)')
axs1.set_title("Sine y = sin(x)")
axs1.set_ylabel("Y")
axs1.grid(True)

axs2 = fig.add_subplot(gs[0, 1])
axs2.plot(x, z, color='red', label='cos(x)')
axs2.set_title("Sine z = cos(x)")
axs2.set_ylabel("Y")
axs2.grid(True)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
print(titanic.shape)
```

```python
print(titanic.describe
      )
```

```python
titanic.dropna()
```

```python
titanic.isnull().sum()
```

```python
sns.countplot(x='survived', data = titanic)
plt.xlabel='Survived'
plt.ylabel='Count'
plt.title('Survived Count')
plt.show()
```

```python
plt.hist(titanic['age'], bins=10)
plt.xlabel="Age"
plt.ylabel="Count"
plt.title('Age Distribution')
plt.show()
```

```python
plt.scatter(titanic['age'], titanic['fare'])
plt.xlabel='Age'
plt.ylabel = 'Fare'
plt.title('Age vs Fare')
plt.show()
```

```python
sns.boxplot(x=titanic['survived'], y=titanic['fare'])
plt.xlabel('Survival Status')
plt.ylabel('Fare')
plt.title('Survival Status vs. Fare')
plt.show()
```

```python
# Correlation analysis
correlation = titanic[['age', 'fare']].corr()
print(correlation)
```

```python
# Cross-tabulation
cross_tab = pd.crosstab(titanic['pclass'], titanic['survived'])
print(cross_tab)
```

```python
sns.catplot(x= 'sex', hue = 'survived', kind = 'count', data= titanic)
```

```python
group = titanic.groupby(['pclass', 'survived'])
pclass_survived = group.size().unstack()
sns.heatmap(pclass_survived, annot= True, fmt= 'd')
```

```python
sns.violinplot(x='sex', y='age', hue ='survived', data= titanic , split= True)
```

```python
sns.catplot(x='embarked', hue = 'survived', kind = 'count' ,col='pclass' , data = titanic)
```

```python
plt.figure(figsize=(10,6))
# sns.heatmap(titanic.corr(), annot= True)
sns.histplot(data=titanic, x='age', kde=True, hue='sex')
plt.title('Age Distribution by Sex')
plt.xlabel="age"
plt.ylabel='distribution'
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic, x='class', y='survived', hue='sex')
plt.title('Survival Rate by Passenger Class and Gender')
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.countplot(data=titanic, x='embarked', hue='survived')
plt.title('Survival Count based on Embarkation Port')
plt.show()
```

```python
import numpy as np
import numpy.linalg as lin
```

```python
x = np.array([
    [15, 30, 3, 12, 15, 1],
    [12, 25, 2, 10, 8, 1],
    [18, 35, 4, 16, 12, 1],
    [8, 5, 1, 5, 2, 1],
    [16, 4, 3, 14, 4, 1],
    [12, 3, 2, 8, 5, 1]
])

y = np.array([[50, 35, 65, 25, 55, 25]])

w = np.array(['w1', 'w2', 'w3', 'w4', 'w5', 'b'])
```

```python
x_det = lin.det(x)

x_det
```

```python
x_inv = lin.inv(x)
x_inv
```

```python
x_inv.shape
```

```python
y = y.T
```

```python
w_val = np.matmul(x_inv , y)
```

```python
w_val
```

```python
import matplotlib.pyplot as plt

plt.plot(w, w_val)
plt.show()
```

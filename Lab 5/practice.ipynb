# SIGMOID FUNCTION

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Synthetic data (Independent variable)
x = np.linspace(-10, 10, 100)

# Appling sigmoid function to x
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid function")
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

# TASK 01

```python
def area(x1, y1, x2, y2):
    return max(0, (x2 - x1) * max(0, (y2 - y1)))

def iou(rectA, rectB):
    x1_A, y1_A, x2_A, y2_A = rectA
    x1_B, y1_B, x2_B, y2_B = rectB

    #intersection 
    x1_inter = max(x1_A, x1_B)
    y1_inter = max(y1_A, y1_B)
    x2_inter = min(x2_A, x2_B)
    y2_inter = min(y2_A, y2_B)

    # area of intersection
    inter_area = area(x1_inter, y1_inter, x2_inter, y2_inter)

    # area
    area_A = area(x1_A, y1_A, x2_A, y2_A)
    area_B = area(x1_B, y1_B, x2_B, y2_B)

    # Area of Union
    union_area = area_A + area_B - inter_area

    #IoU
    iou_value = inter_area / union_area if union_area != 0 else 0
    return iou_value

rectA = (1, 1, 3, 3) 
rectB = (2, 2, 4, 4)

print(f"IoU: {iou(rectA, rectB)}")
```

# TASK 03

```python
def visualize(rectA, rectB):
    fig, axs = plt.subplots()

    rect1 = patches.Rectangle((rectA[0], rectA[1]), rectA[2] - rectA[0], rectA[3] - rectA[1], linewidth=2, edgecolor='r', facecolor='none')
    rect2 = patches.Rectangle((rectB[0], rectB[1]), rectB[2] - rectB[0], rectB[3] - rectB[1], linewidth=2, edgecolor='b', facecolor='none')

    axs.add_patch(rect1)
    axs.add_patch(rect2)

    x1_inter = max(rectA[0], rectB[0])
    y1_inter = max(rectA[1], rectB[1])
    x2_inter = min(rectA[2], rectB[2])
    y2_inter = min(rectA[3], rectB[3])

    if x2_inter > x1_inter and y2_inter > y1_inter:
        inter_rect = patches.Rectangle((x1_inter, y1_inter), x2_inter - x1_inter, y2_inter - y1_inter, linewidth=2, edgecolor='g', facecolor='green', alpha=0.3)
        axs.add_patch(inter_rect)

    axs.set_xlim(0, 10)
    axs.set_ylim(0, 10)

    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    
rectA = (1, 1, 3, 3) 
rectB = (2, 2, 4, 4)

visualize(rectA, rectB)

```

```python
import numpy as np
```

```python
arr = np.random.randint(1, 101, 24)
```

```python
x = np.arange(0, 24)
```

```python
len(arr)
```

```python
subj_score = arr.reshape(6, 4)
subj_score
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, subj_score)
plt.show()
```

```python
total_score = np.sum(subj_score, axis=1)
total_score = total_score.reshape(6, 1)
```

```python
total_score
```

```python
total_grades = np.append(subj_score, total_score, axis=1)
total_grades
```

```python
np.mean(total_score)
```

```python
np.var(total_score)
```

```python
np.std(total_score)
```

```python
i = np.argmin(total_score)
total_score[i]
```

```python
i = np.argmax(total_score)
total_score[i]
```

```python
total_grades[0:2]
```

```python
total_grades = np.delete(total_grades, [0, 1], 0)
```

```python
new_students = np.random.randint(50, 61, 12)
new_students = new_students.reshape(3, 4)
```

```python
total_grad = np.sum(new_students, axis=1)
total_grad = total_grad.reshape(3, 1)
```

```python
total_grad
```

```python
new_students_grades = np.append(new_students, total_grad, axis=1)

new_students_grades
```

```python
final_score = np.append(total_grades, new_students_grades, axis=0)
```

```python
final_score
```

```python
mean = np.mean(final_score, axis=0)
mean.round(2)
```
